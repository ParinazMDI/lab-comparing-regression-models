```python
def numerical_type(data,threshold = 10):
    continous_columns = []
    discrete_columns = []
    
    for column in data.columns:
        if data[column].nunique() > threshold:
            continous_columns.append(column)
        else:
            discrete_columns.append(column)
    continous_df=data[continous_columns]
    discrete_df=data[discrete_columns]
    
    return continous_df,discrete_df
```

```python
continuous_df, discrete_df = numerical_type(numerical_df)
```

```python
#I decided to drop outliers
def remove_outliers_quantile(df,columns):
    for column in columns:
        Q1=df[column].quantile(0.25)
        Q3=df[column].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound= Q1 - 1.5* IQR
        upper_bound = Q3 +1.5 * IQR
        df_clean=df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    
    return df_clean
```

```python
# numerical_df= numerical_df.[numerical_df['customer_lifetime_value','monthly_premium_auto']].remove_outliers_quantile(numerical_df,'customer_lifetime_value','monthly_premium_auto')
# numerical_df
numerical_df = remove_outliers_quantile(numerical_df, ['customer_lifetime_value', 'monthly_premium_auto'])
numerical_df
```

# Check number of unique values in each column, can they be combined in any way to ease encoding? 

```python
#I decided to group some values together
def cleanOperation(x):  

    if 'Two-Door Car' in x:
        return 'Sports Car'
    elif 'Four-Door Car' in x:
        return 'SUV'
    else:
        return 'Luxury SUV'

categoricals_df['vehicle_class'] = list(map(cleanOperation, categoricals_df['vehicle_class']))
categoricals_df['vehicle_class'].value_counts()
```

```python
categoricals_df["renew_offer_type"] = categoricals_df["renew_offer_type"].str.extract('(\d+)').astype(object)
categoricals_df["renew_offer_type"].value_counts()

categoricals_df.head()
```

```python
categoricals_df.education = np.where(categoricals_df['education']=='High School or Below','College',categoricals_df.education)   
categoricals_df.education.value_counts(dropna=False)
```

```python
mapping = {"College " : 0, "Bachelor" : .5, "Master" : 1, "Doctor":1}


categoricals_df["education"] =categoricals_df["education"].replace(mapping)
```

```python
categoricals_df["employmentstatus"] = categoricals_df["employmentstatus"].map({"Medical Leave" : 0, "Disabled" : 0, "Unemployed" : 0, "Retired" : .5, "Employed":2})
```

```python
categoricals_df["location_code"] =categoricals_df["location_code"].map({"Rural" : 0, "Suburban" :.5, "Urban" : 1,})
```

```python
categoricals_df["vehicle_size"] =categoricals_df["vehicle_size"].map({"Small" : 0, "Medsize" :.5, "Large" : 1,})
```

```python
mapping = {'Basic': 0, 'Extended':.5, 'Premium': 1}

# Use map to convert 'coverage' to numerical values
categoricals_df['coverage'] = categoricals_df['coverage'].replace(mapping)
categoricals_df['coverage'].value_counts(dropna=False)
```

```python
#return the date column
categoricals_df['effective_to_date'] = customer_df['effective_to_date']
categoricals_df.head()
```

```python
categoricals_df['effective_to_date'] = categoricals_df['effective_to_date'].astype(str)
date_split = categoricals_df['effective_to_date'].str.split('-', expand=True)
# Convert the components to integers
categoricals_df['Year'] = date_split[0]
categoricals_df['Month'] = date_split[1]
categoricals_df['Day'] = date_split[2]
```
